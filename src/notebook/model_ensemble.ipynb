## Data Preprocessing

```python
df_model = df.copy()

target = "cardiometabolic_risk_state"

cat_cols = [
    "sex",
    "smoking_status",
    "family_history_cvd",
    "fitness_level",
    "activity_type",
]

# Convert to category dtype
for c in cat_cols:
    df_model[c] = df_model[c].astype("category")

# Sort for time-based operations
df_model = df_model.sort_values(["user_id", "day_index"])
```

## Feature Engineering

### Baselines and Deviations

```python
baseline_cols = [
    "resting_hr",
    "hrv",
    "steps",
    "sleep_hours",
    "bp_systolic",
    "fasting_glucose",
]

for col in baseline_cols:
    base = df_model.groupby("user_id")[col].transform("median")
    df_model[f"{col}_baseline"] = base
    df_model[f"{col}_delta"] = df_model[col] - base
```

### Rolling Features

```python
# Weekly 
window = 7

for col in ["resting_hr", "hrv", "steps", "sleep_hours"]:
    df_model[f"{col}_roll_mean"] = (
        df_model.groupby("user_id")[col]
        .transform(lambda x: x.rolling(window, min_periods=1).mean())
    )

    df_model[f"{col}_roll_std"] = (
        df_model.groupby("user_id")[col]
        .transform(lambda x: x.rolling(window, min_periods=1).std())
    )
```

### Data Completeness

```python
missing_flags = [
    "sleep_data_available",
    "bp_measured",
    "glucose_measured",
]

df_model["data_completeness"] = df_model[missing_flags].sum(axis=1)
```

### Drop Non-predictive Columns

```python
drop_cols = ["user_id", "date"]

X = df_model.drop(columns=drop_cols + [target])
y = df_model[target]

groups = df_model["user_id"]
```

## Model

```python
from sklearn.model_selection import GroupKFold
from sklearn.metrics import f1_score
import lightgbm as lgb
import xgboost as xgb
from catboost import CatBoostClassifier
import numpy as np

gkf = GroupKFold(n_splits=5)

lgb_preds = np.zeros((len(X), y.nunique()))
xgb_preds = np.zeros_like(lgb_preds)
cat_preds = np.zeros_like(lgb_preds)

cat_features = [X.columns.get_loc(col) for col in cat_cols]

# LightGBM
for train_idx, val_idx in gkf.split(X, y, groups):

    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    model = lgb.LGBMClassifier(
        n_estimators=500,
        learning_rate=0.05,
        objective="multiclass",
        num_class=y.nunique(),
        random_state=42
    )

    model.fit(X_train, y_train)

    lgb_preds[val_idx] = model.predict_proba(X_val)

# XGBoost
    model_xgb = xgb.XGBClassifier(
        n_estimators=500,
        learning_rate=0.05,
        objective="multi:softprob",
        num_class=y.nunique(),
        tree_method="hist",
        enable_categorical=True,
        random_state=42
    )

    model_xgb.fit(X_train, y_train)

    xgb_preds[val_idx] = model_xgb.predict_proba(X_val)

#CatBoost
    model_cat = CatBoostClassifier(
    iterations=500,
    learning_rate=0.05,
    loss_function="MultiClass",
    verbose=False,
    random_seed=42
    )
    
    model_cat.fit(
    X_train,
    y_train,
    cat_features=cat_features
    )
    cat_preds[val_idx] = model_cat.predict_proba(X_val)
```

### F1 Score and Ensembling the 3 models

```python
final_preds = (lgb_preds + xgb_preds + cat_preds) / 3

y_pred = final_preds.argmax(axis=1)

print("Macro F1:", f1_score(y, y_pred, average="macro"))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y, y_pred))
```

```python
results = df_model[["user_id", "date"]].copy()
results["predicted_risk"] = y_pred
results["confidence"] = final_preds.max(axis=1)

results.head(10)
```
