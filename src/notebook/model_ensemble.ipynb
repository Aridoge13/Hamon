## Data Types

```python
print("\n--- Data Types ---")
display(df.dtypes)

print("\n--- Summary Statistics ---")
display(df.describe())
```

```python
print("\nUnique users:", df["user_id"].nunique())
print("Days per user (should be constant):")
display(df.groupby("user_id")["day_index"].count().describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Missing values

missing = df.isnull().mean().sort_values(ascending=False)

print("\n--- Missingness (fraction) ---")
display(missing)

plt.figure(figsize=(10, 5))
missing[missing > 0].plot(kind="bar")
plt.title("Missing Data Fraction per Column")
plt.ylabel("Fraction Missing")
plt.show()
```

```python
# Target Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x="cardiometabolic_risk_state", data=df)
plt.title("Target Distribution")
plt.show()

print("\nTarget proportions:")
display(df["cardiometabolic_risk_state"].value_counts(normalize=True))
```

```python
target = "cardiometabolic_risk_state"
num_cols = [
    "age",
    "bmi",
    "steps",
    "resting_hr",
    "hrv",
    "sleep_hours",
    "bp_systolic",
    "fasting_glucose",
    "spo2",
    "fatigue_score",
]

plt.figure(figsize=(16, 12))

for i, col in enumerate(num_cols, 1):
    plt.subplot(4, 3, i)
    sns.boxplot(x=target, y=col, data=df)
    plt.title(f"{col} vs Target")

plt.tight_layout()
plt.show()
```

```python
group_means = df.groupby(target)[num_cols].mean()

group_means.T.plot(kind="bar", figsize=(14, 8))
plt.title("Mean Feature Values by Risk State")
plt.ylabel("Mean Value")
plt.show()
```

## Correlation Heatmap

```python
plt.figure(figsize=(12, 10))
corr = df[num_cols].corr()
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()
```

## Categorical VS Target

```python
cat_cols = [
    "sex",
    "smoking_status",
    "family_history_cvd",
    "fitness_level",
    "activity_type",
]

for col in cat_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=col, hue=target, data=df)
    plt.title(f"{col} vs Target")
    plt.show()
```

```python
for col in cat_cols:
    prop = (
        df.groupby(col)[target]
        .value_counts(normalize=True)
        .unstack()
    )

    prop.plot(kind="bar", stacked=True, figsize=(8, 4))
    plt.title(f"Target Proportions by {col}")
    plt.ylabel("Proportion")
    plt.show()
```

# Numerical vs Categorical

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x="activity_type", y="steps", data=df)
plt.xticks(rotation=45)
plt.title("Steps by Activity Type")
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 4))

example_user = df["user_id"].sample(1).iloc[0]; user_df = df[df["user_id"] == example_user].sort_values("day_index")

# Steps on primary axis
ax1.plot(user_df["day_index"], user_df["steps"], color="tab:orange", label="Steps")
ax1.set_xlabel("Day")
ax1.set_ylabel("Steps", color="tab:orange")
ax1.tick_params(axis="y", labelcolor="tab:orange")

# RHR on secondary axis
ax2 = ax1.twinx()
ax2.plot(user_df["day_index"], user_df["resting_hr"], color="tab:blue", label="RHR")
ax2.set_ylabel("Resting HR", color="tab:blue")
ax2.tick_params(axis="y", labelcolor="tab:blue")

plt.title("Example User Trends")
fig.tight_layout()
plt.show()
```

```python
missing_cols = [
    "sleep_data_available",
    "bp_measured",
    "glucose_measured",
]

for col in missing_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, hue=target, data=df)
    plt.title(f"{col} vs Target")
    plt.show()
```

```python
from sklearn.feature_selection import f_classif

X = df[num_cols].fillna(df[num_cols].median())
y = df[target]

f_vals, _ = f_classif(X, y)

feat_imp = pd.Series(f_vals, index=num_cols).sort_values(ascending=False)

feat_imp.plot(kind="bar", figsize=(10, 5))
plt.title("Feature Signal Strength (ANOVA F-score)")
plt.show()
```
